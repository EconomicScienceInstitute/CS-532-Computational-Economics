# Warm Up:

If I flipped a fair coin 1,000 times how many times would I get heads? How about tails? We could do math, but we're data scientists, so let's just get the computer to do it.

```python
import random
def coinFlip(trials):
    heads, tails = 0,0
    for i in range(trials):
        r=random.randint(0,1)
        heads += r
        tails += 1-r
    return [heads,tails]
trials=1000
[h,t] = coinFlip(trials)
print("Out of {} tosses {} heads and {} tails".format(trials,h,t))
```

# Introduction

Probability is the study of how likely events are to occur. Usually we describe the probability of an event x ($P(x)$) as either a percent chance or as a number between 0 and 1. These are equivalent. Take a moment to think of where you've seen someone talk about probability.

We often talk of a percent chance of rain, percent chances of survival, x to y odds for a sports gamble. Although we talk about probability quite often, it can be hard to explain what it actualy means. Take a moment to ask yourself: What does it mean when the TV says there's a 30% chance of rain?

That was a bad example, because meteorologists actually mean that 30% of the geographical region will experience rain during that day; however, if you went along with the thought experiment you might have arrived at a pretty good personal definition of probability. You probably said something along the lines of: "if we had a magic wand that could recreate this day, many many times about 3/10 of the time it will rain."

For the sake of this primer we can think of probability as the ratio between a desired or measured outcome and all possible outcomes. Or as I like to put it

$P(x)=\frac{\text{count of stuff you want}}{\text{count of everything}}$

# Properties of Probabilities

Probabilities have certain properties that can be useful to us. We will be able to avoid using too much calculation if we rely on simulation, but it's still helpful to know these properties to verify our work.

## 0 $\leq$ P(x) $\leq$ 1

If an event is impossible it has a 0% chance of happening or P(X) = 0.

If an event is certain it has a 100% chance of happening or P(x) = 1.

Most events that we will deal with will be somewhere in between. Otherwise it wouldn't be very interesting.

If A and B are mutually exclusive then the probability of A or B occuring is $P(A\cup B) =P(A)+P(B)$ Just remember OR means ADD.

## Definition: Mutually Exclusive
Two events $x_1$ and $x_2$ are mutually exclusive if $x_1$ implies not $x_2$. This means that one event excludes the other. A good intuitive example is a coin toss. If you get heads, you cannot get tails. These are useful because the sum of their probabilities will be the sum of their union. That means:
$P(x_1)+P(x_2) = P(x_1 \cup x_2)$


## $\sum_{x \epsilon A}P(x)=1$

All this means is that the probabilities of all mutually exclusive possible events must add up to one. We can never be certain what is about to happen, but we know at the very least that something will happen. There is a broader way to write this to encompass non-mutually exclusive events, but let's not worry abou that for now
